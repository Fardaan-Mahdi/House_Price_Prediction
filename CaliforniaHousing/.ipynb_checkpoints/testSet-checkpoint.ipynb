```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

##Displaying all columns of dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('housing.csv')
df.head()
```

## OUR MAIN OBJECTIVE HERE IS TO CREATE A TEST DATASET
### But since housing dont have any identifier column so we are going to use row index as the ID


```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing=df.reset_index()  #adds an index column
housing.head()
```

```python
train_set, test_set=split_train_test_by_id(housing,0.2,'index')
```

```python
# This was alternative approact if we are not willing to add additional instances to dataset
# from sklearn.model_selection import train_test_split
# train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

### On observing previous plots and general awareness we can claim that median_income will be an important paramenter for getting median_house_price but since it is a continuous numerical data we need to divide it into income categories to get a better understanding of the plot

```python
housing['income_cat']=pd.cut(housing['median_income'],
                             bins=[0,1.5,3,4.5,6,np.inf],
                            labels=[1,2,3,4,5])
```

```python
housing.head()
```

```python
housing['income_cat'].hist()
```

```python
# Now you are ready to do stratified sampling based on the income category. 
# For this you can use Scikit-Learn’s StratifiedShuffleSplit class
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
```

```python
for train_index,test_index in split.split(housing,housing['income_cat']):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
strat_test_set['income_cat'].value_counts()/len(strat_test_set)
```

```python
strat_train_set['income_cat'].value_counts()/len(strat_train_set)
```

### See how perfectly the strain_train_set and strain_test_set is satisfying the stratified sampling

```python
strat_train_set.head()
```

```python
# Since we successfully done the stratified sampling so now we dont need income_cat

for col_ in (strat_train_set,strat_test_set):
    col_.drop('income_cat',axis=1,inplace=True)
```

```python
strat_test_set.head()
```

## Visualizing Geographical Data

```python
# To get geographical info plotting scatter plot between latitude and longitude

sns.scatterplot(x='longitude', y='latitude',data=housing,alpha=0.1)

#Alternative
# housing.plot(kind='scatter',x='longitude', y='latitude',alpha=0.1)
```

```python
# Now we will try to get better understanding between population and the median_house_price
#The radius of each circle represents the district’s population (option s),
#and the color represents the price (option c)
```

```python
housing.plot(kind='scatter',x='longitude', y='latitude',alpha=0.4,
            s=housing['population']/100,label='population',figsize=(10,7),
            c='median_house_value',cmap=plt.get_cmap('jet'),colorbar=True)
plt.legend()
```

```python
# This image tells you that the housing prices are very much related to the location
# (e.g., close to the ocean) and to the population density
```

```python
corr_matrix['median_house_value'].sort_values(ascending=False)
```
